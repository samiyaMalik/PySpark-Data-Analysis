```python
df.printSchema()
```

```python
df = df.withColumnRenamed("Employee", "Employee_Name")
df.printSchema()
```

```python
df = df.withColumnRenamed("EAge_Increment", "Age++")
df.printSchema()
```

```python
df1 = df.select("Employee_Name", "Age", "City")
df1.show()
```

```python
df.sort("Age").show()
```

```python
from pyspark.sql import functions as F
df.sort(F.desc("Age")).show()
```

```python
def salary_in_k(x):
    return x/1000
from pyspark.sql import types as T
salary_in_k_udf = F.udf(salary_in_k, T.DoubleType())
df = df.withColumn("Salary_in_k", salary_in_k_udf(F.col(("Salary"))))
df.show()
```

```python
df.filter((df.Age >= 30) & (df.Salary_in_k >= 80.0)).show()
```

```python
df.groupBy("Department").agg(F.sum("Salary")).show()
```

```python
df.groupBy("Department").agg(F.sum("Salary"), F.avg("Salary"), F.min("Salary"),F.max("Salary")).show()
```

```python
df.groupBy("Department").agg(F.sum("Salary").alias("Sum_Salary"),
F.avg("Salary").alias("Average_Salary"),
F.min("Salary").alias("Minimum_Salary"),
F.max("Salary").alias("Maximum_Salary")).where(F.col("Minimum_Salary") >= 45000).show()
```

```python
df.groupBy("Department").agg(F.collect_list("Employee_Name").alias("Employees_List")).show()
```

```python
df.registerTempTable("Df_Table")
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer
indexers = [StringIndexer(inputCol = column, outputCol = column + "_idx") for column in ["Department", "City"]]
pipeline = Pipeline(stages = indexers)
indexed = pipeline.fit(df).transform(df)
indexed.show()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.stat import Correlation
vector_col = "corr_features"
corr_cols = list(set(indexed.columns) - set(["Department", "City","Employee_Name"]))
assembler = VectorAssembler(inputCols = corr_cols, outputCol = vector_col)
indexed_vector = assembler.transform(indexed).select(vector_col)

matrix = Correlation.corr(indexed_vector, vector_col)
result = matrix.collect()[0]["pearson({})".format(vector_col)].values
result
```

```python
import pandas as pd
pd.DataFrame(result.reshape(-1, len(corr_cols)), index = corr_cols, columns = corr_cols)
```
